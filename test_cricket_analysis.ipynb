# **2.Given Feature variables**


1. Player = Name of a Player name
2. Span =  total time between starting the test_cricket match and ending the test criket match of a Player
3. Mat =  Total number of Match played by a player
4. Inns =  Total number of Innings played by a player (One Match = 2 innings_)
5. Balls = Total number of Ball throw by a player
6. Runs = Total runs have been given by a bowler
7. Wkts = Total wicket has been taken by a bowler in his carrier
8. BBI = BBI stands for Best Bowling in Innings  for a player
9. BBM = BBM stands for Best Bowling in Match and gives the combined score over 2 or more innings in one match.
10. Ave = The total number of runs divided by the total number of innings in which the batsman was out.
11. Econ = a player's economy rate is the average number of runs they have conceded per over bowled.Economy refers to the runs you concede in each over of your spell. It is the ratio of total runs conceded to the total overs bowled 
12. SR = Batting strike rate (s/r) is defined for a batsman as the average number of runs scored per 100 balls faced. The higher the strike rate, the more effective a batsman is at scoring quickly.
13. 5 = The total number of 5 wickets has taken in a match by a bowler
14. 10 =  The total number of 10 wickets has taken in a Match by a bowler

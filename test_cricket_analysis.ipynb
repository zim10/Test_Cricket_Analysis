# **Explanation of Feature variables**


1. Player = Name of a bowler name
2. Span =  Playing span (difference bewteen Starting year Ending year)
3. Mat =  Total number of Match played by a bowler
4. Inns =  Total number of Innings bowled by a bowler (One Match = 2 innings_)
5. Balls = Total number of Balls bowled by a bowler
6. Runs = Total runs have been conceded by a bowler
7. Wkts = Total wicket has been taken by a bowler in his career
8. BBI = BBI stands for Best Bowling in Innings  for a bowler
9. BBM = BBM stands for Best Bowling in Match and gives the combined score over 2 or more innings in one match.
10. Ave = bowling average ( like average 22.72 means one bowler take one wicket after giving 22.72 runs)
11. Econ = Economy rate means a bowler conceded run by one over (Econ 2.47 means one bowler conceded 2.47 run per over)

12. SR = balls per wicket (SR 55 means one bowler take one wicket after 55 bowled)
13. 5 = The total number of 5 wickets has taken in an Innings by a bowler
14. 10 =  The total number of 10 wickets has taken in a Match by a bowler

# **Some Questions and Answer**

### **Q1.   How to remove the BBI and BBM columns from dataset?**




```python
# use drop method
df1 = df.drop(['BBI','BBM'], axis = 1)
df1.head(5)
```

**Remark:using drop method, it is possible to remove columns**  



### **Q2.   How many players played for ICC?**




```python
# here working with df1 dataset not orginal dataset
# splitting the 'Player' column to get the information about 'ICC' and 'Country
df_player = df1['Player'].str.split("(", expand=True)
# join 'Country' with the main dataset
df1 = pd.concat([df1, df_player], axis=1)
# dropping the 'Player' columns, after spliting , it is no need
df1 = df1.drop('Player', axis=1)

# renaming the column names
df1 = df1.rename(columns={0: 'Player',
                        1: 'Country',
                        '5' : '5_wkts',
                        '10' :  '10_wkts'      })
display(df1.head(5))  
```

```python
# remove the ")" from the 'Country' column
df1['Country'] = df1['Country'].str.replace(")", "")

# rearrange the columns
new_col_sequence = ['Player', 'Span', 'Mat', 'Inns', 'Balls', 'Runs', 'Wkts','Ave','Econ','SR','5_wkts', '10_wkts','Country']
df1 = df1[new_col_sequence]
display(df1.head(5))
```

```python
# Function create to check the icc
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
# call the function 
df1['played_for_ICC'] = df1['Country'].apply(icc_check)

display(df1.head(5))
```

```python
# value_counts() method is used to check the number of unique value presents in the dataset
df1['played_for_ICC'].value_counts()
```

**Remark:** **only 5 playeres are played for the ICC**

### **Q4.Which player(s) had played for the longest period of time?**

```python
# splitting the 'Span' column based on the "-"
df_span = df1['Span'].str.split("-", expand=True)

# conccatinating the new dataframe with the main dataframe
df1 = pd.concat([df1, df_span], axis=1)

# renaming the newly created column names
df1 = df1.rename(columns={0: "start_year",
                        1: "end_year"})

# removing the "Span" column
df1 = df1.drop("Span", axis=1)

display(df1.head(5))
```

```python
# to check the datatypes for finding the span
df1.dtypes
```

**Remark**: start_year and end_year both are string datatypes, it's not possible to find out the difference from this datatypes and need to convert

```python
# convert the object data type into int using astype method 
df1['start_year'] = df1['start_year'].astype('int') 
df1['end_year'] = df1['end_year'].astype('int')

df1.dtypes
```

**Remark:** now start_year and end_year both are int datatypes

```python
# Findout the total year played by a bowler from two columns 
df1['years_played'] = df1['end_year'] - df1['start_year']
# after finding years_played, then no need to keep two column in dataset
df1 = df1.drop(['start_year', "end_year"], axis=1)

```

```python
# sort the years_played column to findout which players had played for the longest period of time.
df1.sort_values('years_played', ascending = False).head(5)
```

**Imran Khan, Pakistani player, had played for the longest period of time and it is 21 years.**
